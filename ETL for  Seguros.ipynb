## Preprocesamiento Rates

```python
df_rates = pd.read_csv("Data/Rate_PUF.csv")
```

```python
df_rates.head()
```

```python
df_rates.info()
```

```python
# Eliminaci√≥n de columnas

df_rates2 = df_rates.drop(
    columns=[
        "BusinessYear", 
        "SourceName", 
        "ImportDate", 
        "RateEffectiveDate", 
        "RateExpirationDate", 
        "RatingAreaId", 
        "Tobacco"
        ])
df_rates2.head()
```

```python
df_rates2.isna().sum()
```

```python
df_rates2["Age"].unique()
```

```python
# Rates individuales
df_rates_individual = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "IndividualRate"]].dropna().reset_index(drop = True).rename({"IndividualRate":"rate"}, axis=1)
df_rates_individual["category"] = "Individual"
print("Filas: ", df_rates_individual.shape[0])
df_rates_individual.head()
```

```python
# Rates IndividualTobaccoRate
df_rates_tobacco = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "IndividualTobaccoRate"]].dropna().reset_index(drop = True).rename({"IndividualTobaccoRate":"rate"}, axis=1)
df_rates_tobacco["category"] = "Tobbacco"
print("Filas: ", df_rates_tobacco.shape[0])
df_rates_tobacco.head()
```

```python
# Rates Couple
df_rates_couple = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "Couple"]].dropna().reset_index(drop = True).rename({"Couple":"rate"}, axis=1)
df_rates_couple["category"] = "Couple"
print("Filas: ", df_rates_couple.shape[0])
df_rates_couple.head()
```

```python
# Rates OneDependent
df_rates_one_d = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "PrimarySubscriberAndOneDependent"]].dropna().reset_index(drop = True).rename({"PrimarySubscriberAndOneDependent":"rate"}, axis=1)
df_rates_one_d["category"] = "OneDependent"
print("Filas: ", df_rates_one_d.shape[0])
df_rates_one_d.head()
```

```python
# Rates TwoDependent
df_rates_two_d = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "PrimarySubscriberAndTwoDependents"]].dropna().reset_index(drop = True).rename({"PrimarySubscriberAndTwoDependents":"rate"}, axis=1)
df_rates_two_d["category"] = "TwoDependent"
print("Filas: ", df_rates_two_d.shape[0])
df_rates_two_d.head()
```

```python
# Rates ThreeDependent
df_rates_3_d = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "PrimarySubscriberAndThreeOrMoreDependents"]].dropna().reset_index(drop = True).rename({"PrimarySubscriberAndThreeOrMoreDependents":"rate"}, axis=1)
df_rates_3_d["category"] = "ThreOrMoreDependent"
print("Filas: ", df_rates_3_d.shape[0])
df_rates_3_d.head()
```

```python
# Rates CoupleAndOneDependent
df_rates_C1_d = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "CoupleAndOneDependent"]].dropna().reset_index(drop = True).rename({"CoupleAndOneDependent":"rate"}, axis=1)
df_rates_C1_d["category"] = "CoupleAndOneDependent"
print("Filas: ", df_rates_C1_d.shape[0])
df_rates_C1_d.head()
```

```python
# Rates CoupleAndTwoDependents
df_rates_C2_d = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "CoupleAndTwoDependents"]].dropna().reset_index(drop = True).rename({"CoupleAndTwoDependents":"rate"}, axis=1)
df_rates_C2_d["category"] = "CoupleAndTwoDependents"
print("Filas: ", df_rates_C2_d.shape[0])
df_rates_C2_d.head()
```

```python
# Rates CoupleAndThreeOrMoreDependents
df_rates_C3_d = df_rates2[["StateCode","IssuerId", "PlanId", "Age", "CoupleAndThreeOrMoreDependents"]].dropna().reset_index(drop = True).rename({"CoupleAndThreeOrMoreDependents":"rate"}, axis=1)
df_rates_C3_d["category"] = "CoupleAndThreeOrMoreDependents"
print("Filas: ", df_rates_C3_d.shape[0])
df_rates_C3_d.head()
```

```python
# union de las tablas

lista = [df_rates_individual, 
         df_rates_tobacco, 
         df_rates_couple, 
         df_rates_one_d, 
         df_rates_two_d, 
         df_rates_3_d, 
         df_rates_C1_d, 
         df_rates_C2_d, 
         df_rates_C3_d]


df_rates3 = pd.concat(lista, ignore_index=True)
df_rates3.shape
```

```python
df_rates3.isna().sum()
```

```python
df_rates3["category"].unique()
```

```python
df_rates3.to_csv("DataSegurosP/Rates.csv", index=False)
```
