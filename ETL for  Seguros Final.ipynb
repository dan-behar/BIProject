```python
import pandas as pd
import numpy as np
```

```python
df_seguros = pd.read_csv("DataSegurosP/seguros.csv").drop_duplicates().rename(columns={"StateCode":"code"})
df_seguros
```

```python
df_seguros.isna().sum()
```

```python
df_seguros["PlanExpirationDate"] = df_seguros["PlanExpirationDate"].fillna("NoDate")
```

```python
df_seguros = df_seguros.dropna()
df_seguros.shape
```

```python
df_seguros.info()
```

```python
df_profiles = pd.read_csv("DataSegurosP/profiles.csv")
id_profiles = df_profiles[["id_profile", "code"]]
```

```python
df_seguros2 = df_seguros.merge(id_profiles, how="left", left_on="code", right_on="code").reset_index().rename(columns={"index":"id_variante"})
```

Orden para las columnas antes de cargarlas en la base de datos

```python
df_seguros2.columns
```

```python
columnas = [
    'id_variante',
    'id_profile',
    'code', 
    'StandardComponentId',
    'PlanId',
    'Age', 
    'Rate', 
    'Rate_category',
    'Income_monthly', 
    'income_anualy', 
    'IssuerMarketPlaceMarketingName', 
    'MarketCoverage', 
    'PlanType',
    'MetalLevel', 
    'QHPNonQHPTypeId', 
    'IsNoticeRequiredForPregnancy',
    'IsReferralRequiredForSpecialist', 
    'ChildOnlyOffering',
    'DiseaseManagementProgramsOffered', 
    'PlanExpirationDate',
    'OutOfCountryCoverage', 
    'SBCHavingaBabyDeductible',
    'SBCHavingDiabetesDeductible', 
    'SBCHavingSimplefractureDeductible',
    'BeginPrimaryCareCostSharingAfterNumberOfVisits',
    'MaximumUnderageDependent', 
    'DependentMaximumAgRule',
    'EssecialsBenfits', 
    'NoEssecialsBenefits'
]
```

```python
df_seguros2 = df_seguros2[columnas]
df_seguros2.info()
```

```python
df_seguros2[["Rate_category"]]
```

# Carga a la base de datos

```python
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://postgres:example@localhost:5433/postgres")
```

```python
df_seguros2.to_sql("seguros", con=engine, index=False,  chunksize = 1000, if_exists='append')
```
