Vamos a dejar la categoría de tabbaco y puesto que Census no tiene información de fumadores dejaremos que esto sea filtrado por el usuario cuando busque su seguro, así si fuma presiona un botón y se le mostrarán todos los seguros que incluyen afumadores.

Modificaremos la columna de Age para que contenga las edades como números y no como rangos

```python
df_rates["Age"].unique()
```

```python
df_rates["Age"] = df_rates["Age"].str.replace("0-14", "14")

df_rates["Age"] = df_rates["Age"].str.replace("64 and over", "64")
df_rates
```

```python
df_rates = df_rates.astype({"Age":"int16"})
df_rates.dtypes
```

Investigamos que las personas en Estados Unidos están dispuestos a pagar entre un 2 y 5 porciento de su salario mensual en seguros, por lo tanto generaremos un columna que suponga que se pagará un 4% en los seguros y que nos indique cuál debe ser el income minimo mensual que deben tener para pagarlo. Y después cuál es el income mínimo anual para relacionarlo con la tabla de los perfiles.

```python
df_rates.info()
```

```python
df_rates["income_monthly"] = df_rates["rate"] / 0.05
df_rates["income_anualy"] = df_rates["income_monthly"] * 12
df_rates.head()
```

### Benefits

```python
df_benefits = pd.read_csv("Data/Benefits_Cost_Sharing_PUF.csv")
df_benefits.head()
```

```python
df_benefits = df_benefits[[
    "StateCode",
    "StandardComponentId",
    "PlanId",
    "BenefitName",
    "IsEHB",
    "IsCovered",
    "QuantLimitOnSvc",
    "LimitQty",
    "LimitUnit"
]]
```

```python
df_benefits.info()
```

```python
df_benefits.isna().sum()
```

Haremos un filtro para eliminar los seguros dentales de esta tabla

```python
df_benefits = df_benefits[df_benefits["StandardComponentId"].isin(NoDental["StandardComponentId"])]
df_benefits
```

```python
df_benefits.shape
```

```python
df_benefits.isna().sum()
```

Haremos un filtro por los seguros que sí están cubiertos por le plan con la variable de IsCovered

```python
df_benefits["IsCovered"].unique()
```

```python
df_benefits = df_benefits[df_benefits["IsCovered"] == "Covered"]
df_benefits.shape
```

Haré una imputación para la columna de essencial

```python
df_benefits["IsEHB"].unique()
```

```python
df_benefits["IsEHB"] = df_benefits["IsEHB"].fillna("No")
df_benefits.head(5)
```

```python
df_benefits["IsCovered"].unique()
```

```python
df_benefits = df_benefits[[
    "StandardComponentId",
    "PlanId",
    "IsEHB"
]]
df_benefits.head()
```

```python
df_benefits_essential = df_benefits[df_benefits["IsEHB"]=="Yes"]
df_benefits_essential
```

```python
df_benefits_NOessential = df_benefits[df_benefits["IsEHB"]=="No"]
df_benefits_NOessential
```

```python
df_essential_count = df_benefits_essential.groupby(["StandardComponentId", 
                                                    "PlanId"]).size().reset_index(name='counts').rename(columns={"counts":"EssecialsBenfits"})
df_essential_count
```

```python
df_essential_count["EssecialsBenfits"].unique()
```

```python
df_NOessential_count = df_benefits_NOessential.groupby(["StandardComponentId", 
                                                    "PlanId"]).size().reset_index(name='counts').rename(columns={"counts":"NoEssecialsBenefits"})
df_NOessential_count
```

```python
df_benefits2 = df_essential_count.merge(df_NOessential_count, how="inner", left_on="PlanId", right_on="PlanId")
df_benefits2 = df_benefits2[[
    "StandardComponentId_x",
    "PlanId",
    "EssecialsBenfits",
    "NoEssecialsBenefits"
]].rename(columns={"StandardComponentId_x":"StandardComponentId"})

df_benefits2
```
