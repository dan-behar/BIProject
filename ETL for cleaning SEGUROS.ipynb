# Union de las tablas


Vamos a filtrar las tablas por las caraterísticas de los perfiles de cada estados

```python
df_profiles = pd.read_csv("DataSegurosP/profiles.csv").sort_values(by="id_profile")
df_profiles
```

Filtraremos la tabla de rates

```python
state_codes = df_profiles["code"].unique()
state_codes
```

```python
def filtro(df_rates, df_profiles, state_c):

    # Obtención de rangod de Edad y de Income de los salarios
    age_min = df_profiles[df_profiles["code"] == state_c]["age_min"].values[0]
    age_min

    age_max = df_profiles[df_profiles["code"] == state_c]["age_max"].values[0]
    age_max

    income_min = df_profiles[df_profiles["code"] == state_c]["income_min"].values[0]
    income_min

    income_max = df_profiles[df_profiles["code"] == state_c]["income_max"].values[0]
    income_max


    # Filtro por edad e Income del perfil del estado
    df_n = df_rates[df_rates["StateCode"] == state_c]
    df_n = df_n[(df_n["Age"] >= age_min) & (df_n["Age"] <= age_max)]
    df_n = df_n[(df_n["income_anualy"] >= income_min) & (df_n["income_anualy"] <= income_max)]

    return df_n
```

```python
df_rates2 = pd.DataFrame()

for state in state_codes: 
    print(state)
    new = filtro(df_rates, df_profiles, state).reset_index(drop=True)
    df_rates2 = pd.concat([df_rates2, new], ignore_index=True)
df_rates2.shape
```

```python
df_rates2.head()
```

Eliminaremos los duplicados antes de hacer los join entre las tablas

```python
df_rates2 = df_rates2.drop_duplicates().reset_index(drop=True)
df_rates2.shape
```

```python
df_benefits2 = df_benefits2.drop_duplicates().reset_index(drop=True)
df_benefits2.shape
```

```python
df_atributes = df_atributes.drop_duplicates().reset_index(drop=True)
df_atributes.shape
```

```python
df_brules = df_brules.drop_duplicates().reset_index(drop=True)
df_brules.shape
```

Haremos un filtro por los StandardComponentId de la tabla de rates 

```python
# Filtro para los seguros
df_filtro = df_rates2["StandardComponentId"].drop_duplicates()
df_filtro.shape
```

```python
df_atributes = df_atributes[df_atributes["StandardComponentId"].isin(df_filtro)]
```

```python
df_benefits2 = df_benefits2[df_benefits2["StandardComponentId"].isin(df_filtro)]
```

```python
df_brules = df_brules[df_brules["StandardComponentId"].isin(df_filtro)]
```

```python
df_benefits2
```

Union de las columnas

```python
df_seguros = df_rates2.merge(df_atributes, how="left", left_on="StandardComponentId", right_on="StandardComponentId")

df_seguros = df_seguros.merge(df_brules, how="left", left_on="StandardComponentId", right_on="StandardComponentId")

df_seguros = df_seguros.merge(df_benefits2, how="left", left_on="PlanId", right_on="PlanId")

df_seguros.drop_duplicates()

df_seguros.shape
```

```python
df_seguros.to_csv("DataSegurosP/seguros.csv", index=False)
```
