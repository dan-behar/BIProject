- NetworkId: indenfica de la red de cuidados del seguro.
- ServiceAreaId: area de servicio
- FormularyId: id de formulario de medicamentos
- IsNewPlan: indica si el seguro es nuevo en el año actual (bool)
- PlanType: tipo de plan (varios valores)
- MetalLevel: es la categoría de covertura del seguro
- DesignType: si el plan es estandarizado para su MetalLevel
- UniquePlanDesign: si el plan tien un unico diseño (bool)
- IssuerActuarialValue: porcentaje de lo que cubre el seguro.

## Rate_PUF

Esta tabla contiene la tarifas de los seguros con base en su fumador, la ubicacion geográfica del asegurado y tarifas a nivel familiar. 

- BusinessYear: elminar*
- StateCode. codigo de estado 
- IssuerId: id de aseguradora 
- SourceName: fuente de datos* 
- ImportDate: fecha de impostacion*
- RateEffectiveDate: fecha de cuando el seguro se hace efectivo*
- RateExpirationDate: fecha de espiración del seguro*
- PlanId: id del plan de seguro. 
- RatingAreaId: areas geográficas dentro del estado, para calificacion.
- Tobacco: si es un fumador*
- Age: rango de edad del suscriptor del seguro o dependientes
- IndividualRate: cobro a individual 
- IndividualTobaccoRate: tarifa a individual que consume tabaco
- Couple: suscriptor y pareja
- PrimarySubscriberAndOneDependent: suscriptor y un dependediente 
- PrimarySubscriberAndTwoDependents: suscriptor y dos dependientes 
- PrimarySubscriberAndThreeOrMoreDependents: suscriptor y tres o más dependientes
- CoupleAndOneDependent: suscriptor, pareja y un dependiente 
- CoupleAndTwoDependents: suscriptor, pareja y dos dependientes
- CoupleAndThreeOrMoreDependent: suscriptor, pareja y tres o más dependientes


Las columnas de IndividualRate en adelante representa la tarifa que se cobra por esa variación del seguro, podemos crear una nueva variable que sea tipo de seguro. Por ejemplo: 

- individual
- individual tabbaco
- Paraje, un dependiente
- dos dependientes
- tres o más dependientes
- pareja y dependetes, etc.
